Use the _author_ endpoint of the [Academic Graph API](https://api.semanticscholar.org/api-docs/graph#tag/Author-Data) to _search_ these researchers in the database based on their names. For each researcher in your list, find:

```python
import requests
import pickle

with open("names.pkl", "rb") as f:
    names = pickle.load(f)
```

```python
def get_collaborator_ids(papers: list):
    ids = []
    for paper in papers:
        for author in paper["authors"]:
            id = author["authorId"]
            if id not in ids:
                ids.append(id)
    return ids
```

```python
author_api_url = "https://api.semanticscholar.org/graph/v1/author/search"
all_author_info = {}
for name in names:
    author_info = {}
    params = {"query": name, "fields": "papers.authors", "limit": 1}
    try:
        response = requests.get(author_api_url, params=params).json()
        data = response["data"][0]
        author_info["author_id"] = data["authorId"]
        collaborator_ids = get_collaborator_ids(data["papers"])
        author_info["collaborator_ids"] = collaborator_ids
        all_author_info[name] = author_info
    except:
        pass
```

```python
unique_author_ids = {}
for author in all_author_info.values():
    if author["author_id"] not in unique_author_ids:
        unique_author_ids[author["author_id"]] = 1
    
    for id in author["collaborator_ids"]:
        if id not in unique_author_ids:
            unique_author_ids[id] = 1
```

```python
batch_url = "https://api.semanticscholar.org/graph/v1/author/batch"
data = {"ids": list(unique_author_ids.keys())[:2]}
params = {"fields": "aliases,papers.title,papers.year,papers.externalIds,papers.s2FieldsOfStudy,papers.citationCount,papers.abstract,name,papers.authors"}
response = requests.post(batch_url, json=data, params=params).json()
```

```python
response
```

```python
def get_citation_count(papers: list) -> int:
    citations = 0
    for paper in papers:
        citations += paper["citationCount"]
    return citations
```

```python
def get_primary_field(papers: list) -> str:
    fields = {}
    for paper in papers:
        fields_of_study = paper["s2FieldsOfStudy"]
        for field in fields_of_study:
            category = field["category"]
            if category not in fields:
                fields[category] = 1
            else:
                fields[category] += 1
    return max(fields, key=fields.get)
```

```python
def get_paper_author_ids(paper: dict) -> list:
    return [a["authorId"] for a in paper["authors"]]
```

```python
author_ids = []
author_names = []
all_author_aliases = []
citation_counts = []
fields = []

for author_info in response: 
    author_ids.append(author_info["authorId"])
    author_names.append(author_info["name"])
    all_author_aliases.append(author_info["aliases"])
    citation_counts.append(get_citation_count(author_info["papers"]))
    fields.append(get_primary_field(author_info["papers"]))
```

```python
paper_ids = []
titles = []
years = []
dois = []
citation_counts = []
all_fields = []
all_author_ids = []

for author_info in response:
    for paper in author_info["papers"]:
        paper_ids.append(paper["paperId"])
        titles.append(paper["title"])
        years.append(paper["year"])
        if "DOI" in paper["externalIds"]:
            dois.append(paper["externalIds"]["DOI"])
        else:
            dois.append("None")
        citation_counts.append(paper["citationCount"])
        all_fields.append([field["category"] for field in paper["s2FieldsOfStudy"]])
        all_author_ids.append(get_paper_author_ids(paper))
```

```python
import pandas as pd
author_dict = {}
author_dict["id"] = author_ids
author_dict["name"] = author_names
author_dict["aliases"] = all_author_aliases
author_dict["citation_count"] = citation_counts
author_dict["field"] = fields

df = pd.DataFrame(author_dict)
df.to_csv("authors.csv")
```
